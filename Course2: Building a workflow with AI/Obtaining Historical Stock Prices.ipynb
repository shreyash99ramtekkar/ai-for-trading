```python
# Get s&p 500 company symbols
s_n_p_stocks = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]['Symbol'].tolist()
```

```python
companies = [ stock for stock in s_n_p_stocks if '.B' not in stock]
companies
```

```python
# down load the data for all s&p 500 companies
start_date = datetime.now() - timedelta(days=365*5)
end_date = datetime.now()
historical_data = yf.download(companies,start=start_date, end=end_date)
historical_data.head()
```

```python
#not losing original data while doing operations
historical_prices = historical_data

# the level 0 columns list

pd.unique(historical_prices.columns.get_level_values(0).to_list())
```
